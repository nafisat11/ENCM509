```python
# Percentage of male and female passengers
gender = [55, 45]

# Counting of people at each age interval shown in age_labels
age = [16, 23, 347, 428, 334, 398, 923, 1015, 216, 11]
age_labels = ['0-9', '10-19', '20-29', '30-39', '40-49', '50-59', '60-69', '70-79', '80-89', '90-99']

fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(11,5))
# Age distribution
ax1.set_xlabel("Age")
ax1.set_ylabel("Quantity")
plt.setp(ax1.get_xticklabels(), rotation=45)
ax1.bar(age_labels, age)
ax1.plot(age, '--ro')

# Gender distribution
ax2.set_title("Gender")
ax2.pie(np.array(gender), labels=["Male", "Female"], autopct='%1.0f%%')
```

- (2 marks) **Exercise 4**


```python
import scipy.stats

array = []
for index, population in enumerate(age):
    array.append(np.full(shape=population, fill_value=index, dtype=np.int))

vector = np.hstack(array)

mean = np.mean(vector)
std = np.std(vector)

x = np.linspace(0, 10.5, len(vector))
y = scipy.stats.norm.pdf(x, mean, std)

plt.plot(x, y)
```
