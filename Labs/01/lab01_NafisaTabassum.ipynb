```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-6, 6)
y = np.sin(x) + np.sin(2*x)/2 + np.sin(3*x)/3 + np.sin(4*x)/4

# Create the figure and axes objects
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12,5))

# Plot the data
ax1.plot(x, y, '-b^')
ax2.plot(x, y, '-.ro')

# Show the grid lines as dark grey lines
ax1.grid(b=True, which='major', color='#666666', linestyle='-')
ax2.grid(b=True, which='major', color='#666666', linestyle='-')

# Setting axes and title
ax1.set_xlabel("x")
ax1.set_ylabel("f(x)")
ax1.set_title("Plot of f(x)")
ax2.set_xlabel("x")
ax2.set_ylabel("f(x)")
ax2.set_title("Plot of f(x)")
```

These are two plots of differing colours, line styles, and markers for the function: $$f(x) = \sin(x) + \frac{sin(2x)}{2} + \frac{sin(3x)}{3} + \frac{sin(4x)}{4}$$

On the left, the plot uses a blue solid line style with triangle_up markers. On the right, the plot uses a red dash-dot line style with circle markers.

```python
import numpy as np
import matplotlib.pyplot as plt

# Percentage of male and female passengers
gender = [55, 45]

# Counting of people at each age interval shown in age_labels
age = [16, 23, 347, 428, 334, 398, 923, 1015, 216, 11]
age_labels = ['0-9', '10-19', '20-29', '30-39', '40-49', '50-59', '60-69', '70-79', '80-89', '90-99']

# Creating a set of subplots
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(11,5))

# Age distribution plot
ax1.set_xlabel("Age", fontweight="bold")
ax1.set_ylabel("Quantity", fontweight="bold")
plt.setp(ax1.get_xticklabels(), rotation=45)
ax1.bar(age_labels, age)
ax1.plot(age, '--ro')

# Gender distribution plot
ax2.set_title("Gender", fontweight="bold")
ax2.pie(np.array(gender), labels=["Male", "Female"], autopct='%1.0f%%', explode=[0.02, 0.02], shadow=True)
```

The ship population is mostly male. The age interval of 70-79 is the highest population while the age interval of 90-99 is the lowest.

- (2 marks) **Exercise 4**: the code and the output generated, including a  brief description, equtions, graphs and conclusion.


```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats

array = []

# Iterating through age intervals to populate count of each index in a list
for index, population in enumerate(age):
    array.append(np.full(shape=population, fill_value=index, dtype=np.int))

# Flattening the nested list into a 1-D vector
vector = np.hstack(array)

# Computing the mean and standard deviation
mean = np.mean(vector)
std = np.std(vector)

# Plotting the normal distribution
x = np.linspace(0, 10.5, len(vector))
y = scipy.stats.norm.pdf(x, mean, std)
plt.plot(x, y)

# Setting axes and title
plt.xlabel("Mean")
plt.ylabel("Probability")
plt.title("Normal Distribution")
```

The mean is 5.335 and the standard deviation is 1.844 of the age of the ship population. This means that the average age onboard the ship is around 50-59 years old.
